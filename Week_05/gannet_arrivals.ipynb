## Import modules

```python
import pandas as pd
import seaborn as sns

# import pearson correlation function 
from scipy.stats import pearsonr
```

## Read in the data and print it out

```python
gannets = pd.read_csv('../Datasets/gannets.csv', index_col=0)
gannets
```

## Plot the data to see what it looks like

```python
ax = sns.scatterplot(data=gannets, x='male', y='female')
ax.plot([10, 30], [10, 30], ls=':')
ax.set_xlabel('Male gannet day of arrival')
ax.set_ylabel('Female gannet day of arrival')
ax.set_aspect('equal')
sns.despine()
```

## Biological, null and alternative hypotheses

- The biological hypothesis is that one of a pair of breeding gannets arrives a few days after its mate.
- The null hypothesis is that arrival times are uncorrelated, i.e.,  = 0. 
- The alternative hypothesis is that arrival times are correlated, i.e.,  $\ne$ 0. 

## Calculate the correlation coefficient

```python
rtest = pearsonr( gannets['female'], gannets['male'] )

print( f'r = {rtest.statistic:.2f}' )
```

## Calculate the 95% CI

```python
ci = rtest.confidence_interval()

print(f'lower bound of 95%CI = {ci.low:.2f}')
print(f'upper bound of 95%CI = {ci.high:.2f}')
```

## Test if the correlation coefficient is significantly different from zero

```python
print( f'p = {rtest.pvalue:.2f}' )
```

## Report the outcome

There was no evidence of a correlation between gannet breeding pair arrival times on Bass Rock (*n* = 10, *r* = 0.46, 95% CI = (0.24, 0.85), *p* = 0.18).
